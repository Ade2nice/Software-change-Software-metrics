```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/elijahadejumo/Downloads/Java/commit.csv')
```

```python
graph= nx.from_pandas_edgelist(df, source="Source", target="Target")
```

```python
nx.number_of_nodes(graph)
```

```python
graph.nodes()
```

```python
most_influentialClass = nx.degree_centrality(graph)
for w in sorted (most_influentialClass, key=most_influentialClass.get, reverse=True):
    print(most_influentialClass[w])
```

```python
most_important = nx.eigenvector_centrality(graph)
for w in sorted (most_important, key=most_important.get, reverse=True):
    print(most_important[w])
```

```python
most_close = nx.closeness_centrality(graph)
for w in sorted (most_close, key=most_close.get, reverse=True):
    print(most_close[w])
```
