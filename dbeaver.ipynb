```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the csv file from computer

```python
df = pd.read_csv(r'/Users/elijahadejumo/Downloads/dbeaver/commit.csv')
```

Convert to an Edgelist

```python
graph= nx.from_pandas_edgelist(df, source="Source", target="Target")
```

Getting the Graph Information

```python
nx.info(graph)
```

Output all the nodes(classes) in graph

```python
graph.nodes()
```

Calculating the Degree_Centrality of the Graph

```python
most_influentialClass = nx.degree_centrality(graph)

    
for w in sorted(most_influentialClass, key=most_influentialClass.get, reverse=True):
    # Check if the value is a numerical type before printing
    if isinstance(most_influentialClass[w], (int, float)):
        print(most_influentialClass[w])
```

```python
most_influentialClass = nx.betweenness_centrality(graph)
for w in sorted (most_influentialClass, key=most_influentialClass.get, reverse=True):
    print(most_influentialClass[w])
```

```python
most_influentialClass = nx.degree_centrality(graph)
for w in sorted (most_influentialClass, key=most_influentialClass.get, reverse=True):
    print(w, most_influentialClass[w])
```

Eigenvector Centrality

```python
most_important = nx.eigenvector_centrality(graph)
for w in sorted (most_important, key=most_important.get, reverse=True):
    print(most_important[w])
```

Closeness Centrality

```python
most_close = nx.closeness_centrality(graph)
for w in sorted (most_close, key=most_close.get, reverse=True):
    print(most_close[w])
```

```python
# Calculate degree centrality
most_influentialClass = nx.degree_centrality(graph)

# Create a graph with node sizes proportional to their degree centrality
pos = nx.spring_layout(graph)  # You can choose other layouts depending on your preference
nx.draw(graph, pos, node_size=[v * 1000 for v in most_influentialClass.values()], with_labels=True)

# Show the plot
plt.show()
```
